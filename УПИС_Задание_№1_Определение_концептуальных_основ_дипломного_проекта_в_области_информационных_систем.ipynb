Для кого предназначено разрабатываемое Вами решение и какие потребности целевой аудитории оно призвано удовлетворить?

«Витрина проектов» — предназначена для всех категорий пользователей платформы LibSpace: студентов, преподавателей и гостей.

Для гостей: Нет возможности быстро оценить профиль и направления деятельности учебного заведения исходя из реальных студенческих проектов.

#### **3. Методология использования**


Каким образом предполагается практическое применение разрабатываемого решения? Опишите основные сценарии взаимодействия пользователей с разрабатываемым решением.

1.  **Сценарий «Преподаватель» (Администрирование шаблонов):**
    *   Преподаватель заходит в раздел управления шаблонами.
    *   Создает новый шаблон (например, "ВКР магистратура") и определяет фиксированный набор полей (элементов), которые нельзя будет перемещать или удалять при заполнении.
    *   Делится шаблоном со студентом.
    *   После заполнения шаблона студентом, выюирает опубликовать или нет данную работу на общую витрину


2.  **Сценарий «Студент» (Создание карточки):**
    *   Пользователь получает шаблон от преподователя.
    *   Система подгружает соответствующий шаблон с фиксированным порядком полей.
    *   Пользователь заполняет карточку, прикрепляет ссылки на медиа-файлы из LibSpace.
    *   Сохраняет карточку как черновик или отправляет на проверку преаодователю.

3.  **Сценарий «Гость» (Просмотр витрины):**
    *   Пользователь заходит на публичную страницу витрины.
    *   Видит только карточки со статусом «Опубликовано».
    *   Может просматривать детали работ, изучать вложения и темы.

#### **4. Входные данные и их источники**


Какие типы данных будут использоваться в качестве входных данных для вашего решения? В каком формате они будут представлены и из каких источников планируется их получение?

1.  **Мета-данные шаблонов (от преподавателей):**
    *   *Тип:* Название шаблона, порядок элементов.
    *   *Источник:* Интерфейс конструктора шаблонов, сохраняемый в БД.
2.  **Данные карточек проектов (от студентов, преподавателей):**
    *   *Тип:* Текстовые поля, даты, ссылки на медиа (ID файлов из хранилища LibSpace), выпадающие списки.
    *   *Формат:* JSON (при отправке форм) / реляционные данные (в БД).
    *   *Источник:* Формы заполнения карточек на клиентской части.
3.  **Данные о пользователях и правах доступа:**
    *   *Тип:* email, роли (студент/преподаватель и т.д.).

4.  **Медиа-файлы:**
    *   *Тип:* Ссылки на видео, документы, изображения.
    *   *Источник:* Интеграция с модулем «Файлы» LibSpace.

Какие результаты будет формировать Ваше решение на выходе?

3.  **Личный кабинет с историей:** Список черновиков и опубликованных работ для каждого автора.
